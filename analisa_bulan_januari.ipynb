```python
import pandas as pd
import os

# Path ke file CSV
file_path = 'C:/TUGAS DATMING/Target_produksi_januari.csv'  # Path ke file CSV

# Memeriksa apakah file ada
if os.path.exists(file_path):
    data = pd.read_csv(file_path, delimiter=';', on_bad_lines='skip')
    # Menampilkan nama kolom untuk memastikan
    print("Kolom yang tersedia: ", data.columns)
else:
    print(f"File not found: {file_path}")

# Menghapus baris yang mengandung missing values
data_clean = data.dropna()

# Menampilkan beberapa baris pertama data yang sudah dibersihkan
print(data_clean.head())
```

```python
# Menghitung statistik dasar
statistics = data_clean.describe()
print(statistics)

# Mengonversi tanggal ke format datetime jika kolom Tanggal ada
if 'Tanggal' in data_clean.columns:
    data_clean['Tanggal'] = pd.to_datetime(data_clean['Tanggal'], format='%d-%m-%Y')

# Menganalisis korelasi antara fitur (jika ada fitur numerik)
correlation_matrix = data_clean.corr()
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi statistik dasar
if 'Tanggal' in data_clean.columns and 'Produksi' in data_clean.columns:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=data_clean, x='Tanggal', y='Produksi')
    plt.title('Produksi Vest Januari - Maret 2024')
    plt.xlabel('Tanggal')
    plt.ylabel('Jumlah Produksi')
    plt.grid(True)
    plt.show()

# Visualisasi korelasi (jika ada fitur numerik)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
