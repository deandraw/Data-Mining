```python
import pandas as pd
import os

# Path ke file CSV
file_path = 'C:/TUGAS DATMING/Target_Produksi_February.csv'  # Path ke file CSV

# Memeriksa apakah file ada
if os.path.exists(file_path):
    data_february = pd.read_csv(file_path, delimiter=';', on_bad_lines='skip')
    # Menampilkan nama kolom untuk memastikan
    print("Kolom yang tersedia: ", data_february.columns)
else:
    print(f"File not found: {file_path}")

# Menghapus baris yang mengandung missing values
data_clean_february = data_february.dropna()

# Menampilkan beberapa baris pertama data yang sudah dibersihkan
print(data_clean_february.head())
```

```python
# Menghitung statistik dasar
statistics_february = data_clean_february.describe()
print(statistics_february)

# Mengonversi tanggal ke format datetime jika kolom Tanggal ada
if 'Tanggal' in data_clean_february.columns:
    data_clean_february['Tanggal'] = pd.to_datetime(data_clean_february['Tanggal'], format='%d-%m-%Y')

# Menganalisis korelasi antara fitur (jika ada fitur numerik)
correlation_matrix_february = data_clean_february.corr()
print(correlation_matrix_february)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi statistik dasar
if 'Tanggal' in data_clean_february.columns and 'Produksi' in data_clean_february.columns:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=data_clean_february, x='Tanggal', y='Produksi')
    plt.title('Produksi Vest Februari 2024')
    plt.xlabel('Tanggal')
    plt.ylabel('Jumlah Produksi')
    plt.grid(True)
    plt.show()

# Visualisasi korelasi (jika ada fitur numerik)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_february, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix - February')
plt.show()
```
